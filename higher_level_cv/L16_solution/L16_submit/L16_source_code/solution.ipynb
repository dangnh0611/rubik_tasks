```python
from yolo import YOLO
from PIL import Image
import os
import imutils
import cv2
import numpy as np
import tracker
```

# Load YOLO model for coconut detection

```python
yolo = YOLO(score=0.51)
```

```python
def detect_image(img):
    r_image, box_detected = yolo.detect_image(img)
    return r_image,box_detected
```

```python
img=cv2.imread('datasets/images/70.png')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
image = Image.fromarray(img.astype('uint8'), 'RGB')
```

```python
r_image,detections=detect_image(image)
print(detections)
```

```python
def process_video(input_path,output_path,process):
    cap = cv2.VideoCapture(input_path)

    # Define the codec and create VideoWriter object
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    out = cv2.VideoWriter(output_path,fourcc, 50.0, (720,1280))
    i=-1
    track=tracker.Tracker(iou_thresh=0.2,max_frames_to_skip=5,max_trace_length=100,trackIdCount=1)
    
    while(cap.isOpened()):
        ret, frame = cap.read()
        if ret is False:
            break
        new_frame=process(track,frame)
        out.write(new_frame)
        cv2.imshow('video',new_frame)
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
    cap.release()
    out.release()
    cv2.destroyAllWindows()
```

```python
def get_center_y(bbox):
    h = bbox[3] - bbox[1]
    y = bbox[1] + h / 2
    return y
```

```python
TOP=350
BOTTOM=900
```

```python
def process(tracker,frame):
    frame=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    frame = Image.fromarray(frame.astype('uint8'), 'RGB')
    r_image,detections=detect_image(frame)
    
    detections_list=[]
    
    for detect in detections:
        center_y=get_center_y(detect)
        if center_y>TOP and center_y<BOTTOM:
            detections_list.append(detect[:4])
            
    bb_detections=np.array(detections_list)
    ret=tracker.Update(r_image,bb_detections)
    ret=np.array(r_image)
    ret=cv2.cvtColor(ret,cv2.COLOR_RGB2BGR)
    return ret
    
```

```python
process_video('coconut.avi','coconut_output.avi',process)
```
